### Exercise 1
**Task:** Import matplotlib and plot a basic line graph.


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
x=np.arange(11)
print(x)
y=x*2
print(y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,color='blue',linestyle='dashdot',linewidth=1,marker='*')
plt.show()
````

### Exercise 9
**Task:** Plot a simple bar chart.

```python
# Your code here

```


````python
students={'john':25,'Alice':20,'Bob':18,'Diana':32}
names=list(students.keys())
ages=list(students.values())
plt.bar(names,ages,color='pink')
plt.xlabel('Names')
plt.ylabel('Ages')
plt.xticks(rotation=45)  
plt.title('Bar Plot of Students Ages')
plt.show()
````
